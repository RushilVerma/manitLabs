```
#======================================================================
# Author - Dr. Manish Kashyap
#======================================================================

# PURPOSE : 1D time frequency analysis using fft (MAGNITUDE PLOT)


#import my_package.my_functions as mf # This is a user defined package
import cv2
import matplotlib.pyplot as plt
import numpy as np
import scipy.ndimage as sci
import scipy.fft as sfft

#-----------------------------------------------------------------
#        Constructing a sin signal using sampling theorem
#-----------------------------------------------------------------

Fs=1000
T=1/Fs
L=1001   # Keep this odd for ease of Mathematics
t=np.linspace(0,L-1,L)*T
f=20
sig1=np.sin(2*np.pi*f*t)

fig1,ax1=plt.subplots()
fig1.show()
plt.plot(t,sig1,label='sine curve')
plt.grid()
plt.title("Input Signal",fontsize=20)       
plt.xlabel("The Time Axis",fontsize=15)     
plt.ylabel("The Amplitude axis",fontsize=15)
plt.legend()

#-----------------------------------------------------------------
#        Bringing that Signal into Frequency Domain
#-----------------------------------------------------------------

freq_axis=np.linspace(0,L-1,L)*Fs/L
fft_sig1=sfft.fft(sig1)/L
fig2,ax2=plt.subplots()
fig2.show()
plt.plot(freq_axis,np.abs(fft_sig1),label='FFT of Signal (Without FFTSHIFT)')
plt.grid()
plt.title("FFT of Signal (Without FFTSHIFT)",fontsize=20,color='b')       
plt.xlabel("The Frequeny Axis",fontsize=15)     
plt.ylabel("The Magnitude axis",fontsize=15) 
plt.legend()

#-----------------------------------------------------------------
#        Using FFT-SHIFT to correctly display the plot
#-----------------------------------------------------------------

freq_axis=np.linspace(-(L-1)/2,(L-1)/2,L)*Fs/L
fft_sig1=sfft.fftshift(sfft.fft(sig1)/L)
fig3,ax3=plt.subplots()
fig3.show()
plt.plot(freq_axis,np.abs(fft_sig1),label='FFT of Signal (With FFTSHIFT)')
plt.grid()
plt.title("FFT of Signal (With FFTSHIFT)",fontsize=20,color='b')       
plt.xlabel("The Frequeny Axis",fontsize=15)     
plt.ylabel("The Magnitude axis",fontsize=15) 
plt.legend()

print("Completed Successfully ...")





```
